```python
caminho_do_csv = r'D:\Gilberto\Gilberto Pb - Compass\Sprints\Sprint 03\Exercicios\secao_05\actors.csv'

with open(caminho_do_csv, 'r', encoding='utf-8') as arquivo:
    conteudo = arquivo.readlines()[1:]

total_bruto = 0
count = 0

for linha in conteudo:
    dados_linha = []
    campo = ""
    tem_aspas = False

    for caracteres in linha:
        if caracteres == '"':
            tem_aspas = not tem_aspas
        elif caracteres == ',' and not tem_aspas:
            dados_linha.append(campo.strip())
            campo = ""
        else:
            campo += caracteres
    dados_linha.append(campo.strip())

    try:
        bruto = float(dados_linha[-1].replace(',', ''))
        total_bruto += bruto
        count += 1
    except ValueError:
        pass

media_bruta = total_bruto / count if count != 0 else 0

print(f'A média da receita bruta é de {media_bruta:.2f} milhões de dólares')
```
