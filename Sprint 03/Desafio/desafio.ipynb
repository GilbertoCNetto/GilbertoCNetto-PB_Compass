## Parte 1 - Lendo o Dataset e eliminando as linhas duplicadas.

### Importando as bibliotecas e lendo o arquivo 'googleplaystore.csv'.

```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

dataset = pd.read_csv('D:/Gilberto/Gilberto Pb - Compass/Sprints/Sprint 03/Desafio/googleplaystore.csv') # pd.read_csv para ler o arquivo csv
```

### Mostrando o Dataset.

```python
display(dataset) # display é uma variação do print usado no pandas, deixa melhor a visualização na minha opinião
```

### Eliminando linhas duplicadas.

```python
dataset = dataset.drop_duplicates('App') # Eliminei as linhas duplicadas pelos Apps, pois foi o que considerei mais correto
display(dataset)
```

## Parte 2 - Criar o Gráfico de Barras com 5 Apps mais instalados.

### Ordenando por quantia de Instalações.

```python
dataset_ordenado = dataset.sort_values(by='Installs', ascending=False) # sort_values, ordena os valores de acordo com os parâmetros passados, nesse caso pelas instalações e em ordem decrescente
display(dataset_ordenado)
```

### Removendo a linha fora de padrão.

```python
dataset = dataset.drop(index=10472) # drop para remover a linha fora de padrão
```

### Transformando os valores da coluna 'Installs' em tipo numérico e verificando o resultado. 

```python
dataset['Installs'] = dataset['Installs'].str.replace(',', '') # trocando a virgula por espaço vazio pra poder transformar em númerico
dataset['Installs'] = dataset['Installs'].str.replace('+', '') # mesma coisa que fiz com a virgula

dataset['Installs'] = pd.to_numeric(dataset['Installs'], errors='coerce') # pd.to_numeric transforma em númerico

display(dataset['Installs'])
```

### Lista ordenada após a remoção de linhas duplicadas.

```python
dataset_ordenado = dataset.sort_values(by='Installs', ascending=False) 
display(dataset_ordenado)
```

### Criando o Gráfico de barras com os 5 Apps mais instalados.

```python
def formatar_installs(x, _): # Função pra formatar as células do eixo y
    if x >= 1e9:
        return f'{x/1e9:.1f}B+'
    elif x >= 1e6:
        return f'{x/1e6:.1f}M'
    else:
        return str(int(x))
    

plt.style.use('ggplot') # Serve pra estilizar
plt.figure(figsize=(9, 6))
plt.bar(dataset_ordenado.head(5)['App'], dataset_ordenado.head(5)['Installs'], ec="k", color='crimson', alpha=1) # define como gráfico de barras e os parâmetros desejados
plt.title('5 Apps mais instalados')
plt.xlabel('Apps')
plt.ylabel('Instalações')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(formatar_installs)) # Puxa a função de formatação
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

## Parte 3 - Criar o Gráfico de Pizza(pie chart) das categorias por frequência.

### Gráfico de pizza, frequência de categorias de apps.

```python
dataset_pizza = dataset['Category'].value_counts()
instalacoes_minimas = 170 # Fui testando até encontrar um valor satisfatório
others = dataset_pizza[dataset_pizza < instalacoes_minimas].sum()


dataset_pizza = dataset_pizza[dataset_pizza >= instalacoes_minimas]
dataset_pizza['OTHERS'] = others
# Todas as categorias que tenham frequência menor que 170 entraram no 'OTHERS'

plt.style.use('ggplot')
plt.figure(figsize=(10, 8))
explode = [0.05 if valor == dataset_pizza.max() else 0 for valor in dataset_pizza] # usei para destacar a categoria mais frequente
plt.pie(dataset_pizza, labels=dataset_pizza.index, autopct='%1.1f%%', startangle=90, explode=explode, 
        wedgeprops={'edgecolor': 'white'}, pctdistance=0.85, labeldistance=1.075)
plt.title('Distribuição por Categoria')
plt.show()
```

## Parte 4 - Mostrar qual o App mais caro no dataset.

### Verificando o tipo da coluna 'Price'

```python
print(dataset['Price'].dtype) # usei dtype para ver o tipo da coluna 'price'
```

### Transformando a coluna 'Price' em numérica e mostrando o App mais caro.

```python
def converter_preco(preco):
    if isinstance(preco, str):
        preco = preco.replace("$", "").strip()
        try:
            return float(preco)
        except ValueError:
            return None
    else:
        return preco


dataset['Price'] = dataset['Price'].apply(converter_preco)
dataset_ordenado_preco = dataset.sort_values(by='Price', ascending=False)

display(dataset_ordenado_preco.head(1))
```

## Parte 8 - Criar 2 novos gráficos em cima do dataset.

### Gráfico de Barras Horizontal dos Top 10 Apps com mais Reviews

```python
def formatar_reviews(x, _): # Função pra formatar as células do eixo x
    if x >= 1e6:
        return f'{x/1e6:.1f}M+'
    else:
        return str(int(x))
    
dataset_reviews = dataset[['App', 'Reviews']].dropna().sort_values(by='Reviews', ascending=False).head(10) # ordenando a lista, pegando os 10 primeiros e removendo os NaN
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.barh(dataset_reviews['App'], dataset_reviews['Reviews'], ec="k", color='cornflowerblue')
plt.xlabel('Número de Reviews')
plt.title('Top 10 Aplicativos com Mais Reviews')
plt.gca().invert_yaxis()
plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(formatar_reviews))
plt.show()
```

### Gráfico Histograma da Frequência das notas(Rating).
